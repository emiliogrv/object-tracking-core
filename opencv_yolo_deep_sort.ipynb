# Detecting stuff

```python
from detectors.yolo_deep_sort import YOLODeepSort

# source = 'https://motchallenge.net/sequenceVideos/MOT16-13-raw.mp4'
source = 'https://upload.wikimedia.org/wikipedia/commons/c/c6/Traffic_flow_in_front_of_UI_.webm'
# source = 'https://upload.wikimedia.org/wikipedia/commons/thumb/3/32/Hajjiabad%2C_Zeberkhan%2C_Nishapur_-_Pictures_Of_People_02.jpg/1280px-Hajjiabad%2C_Zeberkhan%2C_Nishapur_-_Pictures_Of_People_02.jpg'
detector = YOLODeepSort({"DEBUG": True})

is_video, output_path = detector.detect(source, 'result')
```

```python
from ipywidgets import Video
from IPython.display import Image

if is_video:
  # In some cases, you'll be unable to see the video wi an error like:
  # "no video with supported format and mime type found"
  # so please check your browser compatibility
  # https://www.quirksmode.org/html5/tests/video.html
  output = Video.from_file(output_path)
else:
  output = Image(output_path)

output
```
